## Converting images to required format

```
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
df = train_df.append(test_df)

print(df.shape)
df.head(3)
```

```
len(np.unique(df['filename']))
```

```
df['xmax'] = df['xmax']/1024
df['xmin'] = df['xmin']/1024
df['ymax'] = df['ymax']/1024
df['ymin'] = df['ymin']/1024
df['height'] = (df['ymax'] - df['ymin'])
df['width'] = (df['xmax'] - df['xmin'])

df['x_centre'] = (df['xmin'] + df['width']/2).astype('str')
df['y_centre'] = (df['ymin'] + df['height']/2).astype('str')
df['height'] = df['height'].astype('str')
df['width'] = df['width'].astype('str')
df['class'] = [str(0) for i in range(df.shape[0])]
df = df[['filename','class','x_centre','y_centre','width','height']]
```

```
df.head()
```

```
filenames = df['filename'].unique().tolist()
base_path = 'data/custom/'
label_path_ = 'data/custom/labels'
if not os.path.exists(label_path_):
  os.makedirs(label_path_)

#train_files
train_files = train_df['filename'].unique().tolist()
#train and validation index files
train_txt = 'data/custom/train.txt'
val_txt = 'data/custom/valid.txt'
trainF = open(train_txt, "w")
valF = open(val_txt,"w")

for f in filenames:
  mode = 'train' if f in train_files else 'val'
  image_path = base_path + f'images/{str(f)}'
  ann_name = str(f).split('.')[0] + '.txt'
  annotation_path = base_path + f'labels/{ann_name}'
  
  # appending location of train/validation images to corresponding files
  if mode == 'train':
      trainF.write(image_path)      
      trainF.write("\n") 
  else:
      valF.write(image_path)
      valF.write("\n")

  print(mode, annotation_path)

  # appending class_idx,x_centre,y_centre,w,h
  dff = df[df['filename'] == f]
  outF = open(annotation_path, "w")
  for i in dff.values:
      line = ' '.join(i[1:])
      outF.write(line)      
      outF.write("\n")
  outF.close()


trainF.close()
valF.close()
  
```
